```python
###working

import pandas as pd
import re


regex_patterns = {
    'first_name': r"(\w+) \w+:",
    'last_name': r"\w+ (\w+):",
    'address': r"(?:at|in|from) (\d* [\d\w\s]+ (?:St|Blvd))",
    'city': r'\b(?:in|hub of|streets of) ([A-Z][a-z]*(?:\s[A-Z][a-z]+)*)',
    'county': r"(?:for|in) ([\w\s]+) County",
    'state': r"(?:in|across|from) ([A-Z]{2})\b",
    'phones': r"(\d{3}-\d{3}-\d{4})(.*?(\d{3}-\d{3}-\d{4}))?",
    'email': r"(\w+@\w+\.\w+)"
}

# Company name extraction function
def extract_company_name(description):
    regex_company_name_1 = r"\b[A-Z]\w+, [A-Z]\w+ [A-Z] [A-Z]\w+\b"
    regex_company_name_2 = r"(?:from|of) ([A-Z]\w*\s[A-Z]\w*(?:\s[A-Z]\w*)?)"

    match = re.search(regex_company_name_1, description)
    if match:
        return match.group(0)
    match = re.search(regex_company_name_2, description)
    if match:
        return match.group(1)
    return None

# Function to extract data from the description
def extract_info(description):
    info = {}
    # Handle company name separately
    info['company_name'] = extract_company_name(description)
    # Extract other information
    for key, pattern in regex_patterns.items():
        if key == 'phones':
            match = re.search(pattern, description)
            if match:
                info['phone1'] = match.group(1)
                info['phone2'] = match.group(3) if match.group(3) else None
        else:
            match = re.search(pattern, description)
            info[key] = match.group(1) if match else None
    return info

# Read the CSV data into a DataFrame
df = pd.read_csv('/content/Assignment0PartA.csv')

# Apply the extract_info function to each description and update the DataFrame
for key in ['company_name'] + list(regex_patterns.keys()):
    if key != 'phones':  # Exclude 'phones' key as it's handled separately
        df[key] = df['Description'].apply(lambda desc: extract_info(desc).get(key))

# Separately handling phone numbers
df['phone1'], df['phone2'] = zip(*df['Description'].apply(lambda desc: (extract_info(desc).get('phone1'), extract_info(desc).get('phone2'))))

# Example of how to access the extracted information
print(df.head())

# Save the updated DataFrame back to CSV
df.to_csv('story_updated.csv', index=False)
```

```python
import re
import pandas as pd

# Create a mapping of combined condition names to individual conditions
combined_conditions_mapping = {
    "HIV/AIDS": ["HIV", "AIDS"],
    "Accidents/injuries": ["Accidents", "Injuries","Accident","Injury"],
    "Heart disease": ["Heart disease", "Heart diseases"],
    "Oral health issues": ["Oral health issue", "Oral health issues"],
    "Chronic lower respiratory disease": ["Chronic lower respiratory disease", "Chronic lower respiratory diseases"],
    "Parkinson's disease": ["Parkinson's disease", "Parkinson's diseases","Parkinson's"],
    "Alzheimer's disease": ["Alzheimer's disease", "Alzheimer's diseases","Alzheimer's"],
    "Respiratory diseases": ["Respiratory disease", "Respiratory diseases"],
    "Influenza and Pneumonia": ["Influenza", "Pneumonia"]
}



health_conditions = [
    "Heart disease",
    "Cancer",
    "Stroke",
    "Respiratory diseases",
    "Alzheimer's disease",
    "Diabetes",
    "Influenza and Pneumonia",
    "Kidney disease",
    "Septicaemia",
    "Liver disease",
    "Hypertension",
    "Parkinson's disease",
    "Chronic lower respiratory disease",
    "Accidents/injuries",
    "Osteoporosis",
    "Asthma",
    "Depression",
    "Oral health issues",
    "HIV/AIDS",
    "Tuberculosis",
    "Malaria",
    "Dengue",
    "Hepatitis",
    "Epilepsy",
    "Multiple sclerosis"
]

# Function to count occurrences of health conditions in a volume
def count_health_conditions(volume_filename, health_conditions):
    with open(volume_filename, 'r', encoding='utf-8') as file:
        volume_text = file.read()

    condition_counts = {}

    for condition in health_conditions:
        # Check if it's a combined condition
        if condition in combined_conditions_mapping:
            combined_condition_name = condition
            individual_conditions = combined_conditions_mapping[condition]
            occurrences_combined = 0
#             for ind_condition in individual_conditions:
#                 pattern = re.escape(ind_condition)
#                 occurrences = len(re.findall(pattern, volume_text, re.IGNORECASE))
#                 occurrences_combined += occurrences
            for ind_condition in individual_conditions:
                pattern = re.escape(ind_condition)

                if ind_condition in ["HIV", "AIDS"]:
                    occurrences = len(re.findall(pattern, volume_text))
                else:
                    occurrences = len(re.findall(pattern, volume_text, re.IGNORECASE))
                occurrences_combined += occurrences
            condition_counts[combined_condition_name] = occurrences_combined
        else:
            pattern = re.escape(condition)
            occurrences = len(re.findall(pattern, volume_text, re.IGNORECASE))
            condition_counts[condition] = occurrences

    return condition_counts

# Create a DataFrame to store the results
df = pd.DataFrame(columns=health_conditions)

# Loop through the five volumes and count occurrences for each
for i in range(1, 6):
    volume_filename = f'/content/vol{i}.txt'
    condition_counts = count_health_conditions(volume_filename, health_conditions)
    df = df.append(condition_counts, ignore_index=True)

# Transpose the DataFrame and set volume labels as columns
df = df.transpose()
df.columns = [f'Vol{i}' for i in range(1, 6)]

# Display the formatted DataFrame
# print(df)
# Save the DataFrame to a CSV file
df.to_csv('health_conditions_frequency.csv')

# Print the DataFrame
df
```
