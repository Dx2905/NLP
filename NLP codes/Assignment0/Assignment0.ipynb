## Part A

  For Part A, there're multiple ways of creating the regex patterns. Here we're showing one solution that gives us the answer that we think is comprehensive.


*   For **first_name, last_name, county, state, phone1, phone2, and email**: It's not hard to tell the regex pattern by the context.
*   For **company_name, address, city**: We created the regex pattern after observing the context and came up with solutions that make sense in syntax.





```python
import pandas as pd
import re

# Load the CSV file
csv_path = '/content/Assignment 0 Part A.csv'
data = pd.read_csv(csv_path)

# Updated regular expressions as provided
regex_patterns = {
    'first_name': r'([A-Z][a-z]+)\s[A-Z][a-z]+:',  # Extracting first name
    'last_name': r'[A-Z][a-z]+\s([A-Z][a-z]+):',  # Extracting last name
    'company_name': r'(?:of|from|at)\s+((?:[A-Z][a-z]*,?\s*)+(?:\s+Jr|Esq|Cpa|Service|Dimensions))',  # Extracting company name
    'address': r'(?:at|in|from) (\d+.*?\b(?:St|ZIP|Blvd)\b)',  # Extracting address
    'city': r'\b(?:in|hub of|streets of) ([A-Z][a-z]*(?:\s[A-Z][a-z]+)*)',  # Extracting city
    'county': r'\b([A-Z][a-z]+) County\b',  # Extracting county
    'state': r'\b(A[LKZRAEP]|C[AOT]|DE|FL|GA|HI|I[DLNA]|K[SY]|LA|M[EDAINSOT]|N[EVHJMYCD]|O[HKR]|PA|RI|S[CD]|T[NX]|UT|V[TA]|W[AVIY])\b',  # Extracting state
    'phone1': r'(\d+-\d+-\d+)',  # Extracting first phone number
    'phone2': r'(\d{3}-\d{3}-\d{4})(?!.*\d{3}-\d{3}-\d{4})',  # Extracting second phone number
    'email': r'(\S+@\S+)'  # Extracting email
}

# Apply regular expressions to extract the data
for column, pattern in regex_patterns.items():
    data[column] = data['Description'].apply(lambda x: re.search(pattern, x).group(1) if re.search(pattern, x) else 'N/A')

# Save the updated DataFrame to the same CSV file
data.to_csv('PartA_answer.csv', index=False)

# Display the first few rows of the updated DataFrame
data.head()
```

# Part B

  Our solution for this Part is simple.


1.   Find disease that has combined conditions and create regex pattern for any possible condition that refers to the disease.
2.   For all the disease, we need to try to find the matches by ingoring the uppercase or lowercase(except HIV and AIDS).
3.   5 Volumes of books are in following format: vol1.txt,vol2.txt,vol3.txt,vol4.txt and vol5.txt.



```python
import warnings
warnings.simplefilter("ignore")

# Create a mapping of combined condition names to individual conditions
combined_conditions_mapping = {
    # Each key is a combined condition name and its value is a list of individual conditions
    "HIV/AIDS": ["HIV", "AIDS"],
    "Accidents/injuries": ["Accidents", "Injuries","Accident","Injury"],
    "Heart disease": ["Heart disease", "Heart diseases"],
    "Kidney disease": ["Kidney disease", "Kidney diseases"],
    "Oral health issues": ["Oral health issue", "Oral health issues"],
    "Chronic lower respiratory disease": ["Chronic lower respiratory disease", "Chronic lower respiratory diseases"],
    "Parkinson's disease": ["Parkinson's disease", "Parkinson's diseases","Parkinson's"],
    "Alzheimer's disease": ["Alzheimer's disease", "Alzheimer's diseases","Alzheimer's"],
    "Respiratory diseases": ["Respiratory disease", "Respiratory diseases"],
    "Influenza and Pneumonia": ["Influenza", "Pneumonia"]
}

# List of health conditions to be analyzed

health_conditions = [
    "Heart disease",
    "Cancer",
    "Stroke",
    "Respiratory diseases",
    "Alzheimer's disease",
    "Diabetes",
    "Influenza and Pneumonia",
    "Kidney disease",
    "Septicaemia",
    "Liver disease",
    "Hypertension",
    "Parkinson's disease",
    "Chronic lower respiratory disease",
    "Accidents/injuries",
    "Osteoporosis",
    "Asthma",
    "Depression",
    "Oral health issues",
    "HIV/AIDS",
    "Tuberculosis",
    "Malaria",
    "Dengue",
    "Hepatitis",
    "Epilepsy",
    "Multiple sclerosis"
]

# Function to count occurrences of health conditions in a volume
def count_health_conditions(volume_filename, health_conditions):
    with open(volume_filename, 'r', encoding='utf-8') as file:
        volume_text = file.read()
    # Dictionary to hold the count of each condition
    condition_counts = {}

    for condition in health_conditions:
        # Check if it's a combined condition
        if condition in combined_conditions_mapping:
            combined_condition_name = condition
            individual_conditions = combined_conditions_mapping[condition]
            occurrences_combined = 0
             # Count occurrences for each individual condition
            for ind_condition in individual_conditions:
                pattern = re.escape(ind_condition)
                # Apply case-sensitive matching for specific conditions
                if ind_condition in ["HIV", "AIDS"]:
                    occurrences = len(re.findall(pattern, volume_text))
                else:
                    # Case-insensitive matching for other conditions
                    occurrences = len(re.findall(pattern, volume_text, re.IGNORECASE))  # ignore the letter case when matching the regex pattern
                occurrences_combined += occurrences
            # Store the combined count for the combined condition
            condition_counts[combined_condition_name] = occurrences_combined
        else:
            # Count occurrences for non-combined conditions
            pattern = re.escape(condition)
            occurrences = len(re.findall(pattern, volume_text, re.IGNORECASE))
            condition_counts[condition] = occurrences

    return condition_counts

# Create a DataFrame to store the results
df = pd.DataFrame(columns=health_conditions)

# Loop through the five volumes and count occurrences for each
for i in range(1, 6):
    volume_filename = f'/content/vol{i}.txt'
    condition_counts = count_health_conditions(volume_filename, health_conditions)
    df = df.append(condition_counts, ignore_index=True)

# Transpose the DataFrame and set volume labels as columns
df = df.transpose()
df.columns = [f'Vol{i}' for i in range(1, 6)]

# Save the DataFrame to a CSV file
df.to_csv('health_conditions_frequency.csv')

# Print the DataFrame
df
```
